```python
import random # library for random number generation
import numpy as np # library for vectorized computation
import pandas as pd # library to process data as dataframes
import requests
```

Reading the WebPage and storing the table in a dataframe

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

df_postcode=pd.read_html(url, header=0)[0]

df_postcode.head()
```

Renaming the column 'PostCode' into 'PostalCode'

```python
df_postcode.rename(index=str, columns={"Postcode": "PostalCode"}, inplace=True )
df_postcode.head()
```

Removing the rows which has Borough as 'Not assigned'

```python
df_postcode = df_postcode[df_postcode.Borough != 'Not assigned'].reset_index()
df_postcode
```

Combining the rows which has same PostalCode and updating the Neigbourhood for those rows

```python
foo = lambda a: ",".join(a) 
df=df_postcode.groupby(['PostalCode','Borough']).agg({'Neighbourhood': foo}).reset_index()
```

```python
df
```

Updating the value of 'Neighbourhood' as the vaue of 'Borough' where Neighbourhood is 'Not Assigned'

```python
df.loc[df.Neighbourhood=='Not assigned','Neighbourhood']=df.loc[df.Neighbourhood=='Not assigned','Borough']
```

Reading the location CSV file into a Panda Dataframe

```python
data = pd.read_csv("https://cocl.us/Geospatial_data") 
data
```

Updating the column 'Postal Code' to 'PostalCode' to match the column name in the location dataframe df

Joining the two datasets to get the location details

```python
df_loc = pd.merge(df, data, on='PostalCode', how='left')

df_loc
```
